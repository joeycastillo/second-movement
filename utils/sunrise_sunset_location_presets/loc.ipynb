```python
with open("cities.json", "r", encoding="utf-8") as f:
    loaded_cities = json.load(f)
```

```python
min_population = 1_000_000
min_km = 500

keep = {}
entries_to_remove = []
for city in loaded_cities:
    if city in ["Detroit", "Raleigh"]:
        keep[city] = loaded_cities[city]
    elif "longitude" not in loaded_cities[city] or loaded_cities[city]["longitude"] is None or loaded_cities[city]["population_2025"] < min_population:
        entries_to_remove.append(city)

city_start = None
done = False
while not done:
    if entries_to_remove:
        for entry in entries_to_remove:
            del loaded_cities[entry]
    entries_to_remove = []
    city_names = list(loaded_cities.keys())
    start_index = 0
    if city_start in city_names:
        start_index = city_names.index(city_start)
    for i, city1 in tqdm(enumerate(city_names[start_index:]), desc="Processing cities", total=len(city_names) - start_index):
        info1 = loaded_cities[city1]
        point1 = (info1["latitude"], info1["longitude"])
        for j in range(i+1, len(city_names)):
            city2 = city_names[j]
            if city1 == city2:
                continue
            info2 = loaded_cities[city2]
            if TIMEZONE_MAP[info1["timezone"]] != TIMEZONE_MAP[info2["timezone"]]:
                continue
            point2 = (info2["latitude"], info2["longitude"])
            dist = distance.distance(point1, point2).km
            if dist < min_km:
                entries_to_remove.append(city2)
                if dist < 50:
                    print(f"Remove {city2} for being {round(dist, 0)}km from {city1}")
        if entries_to_remove:
            city_start = city1
            break
        elif city1 == city_names[-1]:
            done = True

loaded_cities.update(keep)
```

```python
continent_order = ["North America", "Asia", "Europe", "Africa", "South America", "Oceania"]
sorted_loaded_cities = dict(sorted(loaded_cities.items()))
max_city_name_len = 6

print("// Data came from here: https://worldpopulationreview.com/cities")
print(f"// All cities have a 2025 population of at least {min_population} and removed all cities that are within {min_km}km of a more populated city that shares a timezone. {len(sorted_loaded_cities)} cities total.")
tz_with_locations = []
for tz_idx in TZ_ORDER:
    found_a_city = False
    for i, continent in enumerate(continent_order):
        for city, content in sorted_loaded_cities.items():
            timezone = TIMEZONE_MAP[content["timezone"]]
            if timezone != tz_idx:
                continue
            if content["continent"] != continent:
                continue
            lat = content["latitude"]
            long = content["longitude"]
            lat = int(round(lat * 100, 0))
            long = int(round(long * 100, 0))
            country = content["country"]
            population = content["population_2025"]
            if not found_a_city:
                found_a_city = True
                tz_with_locations.append(tz_idx)
                print(f"static const sunrise_sunset_long_lat_presets_t {tz_idx}_CITIES[] = {{")
            print(f"    {{ .name = \"{city[:max_city_name_len]}\", .latitude = {lat}, .longitude = {long}, .region = {i} }},  // {city}, {country} (Region: {continent}, 2025 Population: {population})")
    if found_a_city:
        print("};\n")

print("static const sunrise_sunset_timezone_city_group_t sunriseSunsetLongLatPresets[] = {")
print("    { NUM_ZONE_NAMES, NULL, 0 },")
for tz_idx in tz_with_locations:
        print(f"    {{ {tz_idx}, {tz_idx}_CITIES, sizeof({tz_idx}_CITIES)/sizeof({tz_idx}_CITIES[0]) }},")
print("};")



```
